# De opdracht.
Maak een Monte Carlo Simulatie die meerdere keren (duizenden) de competitie speelt met gebruik van deze tabel van kansen. Hoe vaker je een competitie speelt, hoe nauwkeuriger je voorspelling wordt. Houd bij hoeveel punten elk team scoort in de competitie (3 punten voor winst, 1 punt voor gelijk, 0 voor verlies), en maak een overzicht (ranking) van de teams. 

 

Gebruik nu de onderstaande tabel met kansen (Winst, Gelijkspel, Verlies) om een Monte Carlo Simulatie te maken van een competitie, waarin elke ploeg tenminste 1 keer thuis en 1 keer uit tegen dezelfde tegenstander speelt. Uiteraard moeten alle teams tegen elkaar spelen. 

Uitleg ("how to read"): De drie getallen in een veld geven de win-/gelijkspel-/verlieskans aan voor de thuisspelende partij (voor de uitspelende partij is winst en verlies natuurlijk omgedraaid). Bijv. voor de wedstrijd Ajax - PSV, zoek de rij van Ajax in de linkerkolom (thuisspelende partij) en daarbij de juiste kolom met PSV als uitspelend: 54/21/25. Deze waarden betekenen op een volgende dat: de thuisspelende partij (Ajax dus) met 54% kans gaat winnen, met 21% kans gelijk speelt of met 25% kans verliest van de uitspelende partij (PSV dus).

# De code.

### De variabele.
Hier heb ik onder andere de percentages voor elke partij een variabele gegeven, ook heb ik per club alle wedstrijden in een variabele gezet zodat het wat netter lijkt in de code als ik alle wedstrijden nodig heb. De scores worden hier ook aangemaakt, ze beginnen allemaal bij. Er wordt hier ook een DataFrame aangemaakt waar later de vergaarde data in gestopt wordt.

```python
ajax_feyenoord = [65,17,18]
ajax_psv = [54,21,25]
ajax_fcUtrecht = [74,14,12]
ajax_willemII = [78,13,9]

feyenoord_ajax = [30,21,49]
feyenoord_psv = [37,24,39]
feyenoord_fcUtrecht = [51,22,27]
feyenoord_willemII = [60,21,19]

psv_ajax = [39,22,39]
psv_feyenoord = [54,22,24]
psv_fcUtrecht = [62,20,18]
psv_willemII = [62,22,16]


fcUtrecht_ajax = [25,14,61]
fcUtrecht_feyenoord = [37,23,40]
fcUtrecht_psv = [29,24,47]
fcUtrecht_willemII = [52,23,25]

willemII_ajax = [17,18,65]
willemII_feyenoord = [20,26,54]
willemII_psv = [23,24,53]
willemII_fcUtrecht = [37,25,38]

totaal_score_ajax = 0
totaal_score_feyenoord = 0
totaal_score_psv = 0
totaal_score_fcUtrecht = 0
totaal_score_willemII = 0

WIN = 3
DRAW = 1
LOSE = 0

alle_ajax = [ajax_feyenoord, ajax_psv, ajax_fcUtrecht, ajax_willemII, feyenoord_ajax, psv_ajax, fcUtrecht_ajax, willemII_ajax]
alle_feyenoord = [feyenoord_ajax, feyenoord_psv, feyenoord_fcUtrecht, feyenoord_willemII, ajax_feyenoord, psv_feyenoord, fcUtrecht_feyenoord, willemII_feyenoord]
alle_psv = [psv_ajax, psv_feyenoord, psv_fcUtrecht, psv_willemII, ajax_psv, feyenoord_psv, fcUtrecht_psv, willemII_psv]
alle_fcUtrecht = [fcUtrecht_ajax, fcUtrecht_feyenoord, fcUtrecht_psv, fcUtrecht_willemII, ajax_fcUtrecht, feyenoord_fcUtrecht, psv_fcUtrecht, willemII_fcUtrecht]
alle_willemII = [willemII_ajax, willemII_feyenoord, willemII_psv, willemII_fcUtrecht, ajax_willemII, feyenoord_willemII, psv_willemII, fcUtrecht_willemII]

alle_wedstrijden = [alle_ajax, alle_feyenoord, alle_psv, alle_fcUtrecht, alle_willemII]

df = pd.DataFrame(columns = ['Eerste', 'Tweede', 'Derde', 'Vierde', 'Vijfde'])
```

### De 'for loop'.
Dit is het belangrijkste gedeelte van de code, hier wordt elk 'voetbal seizoen' (alle wedstrijden een keer) gesimuleerd, momenteel wordt het seizoen 1000 keer gespeeld. De uitkomst van deze code wordt geplaatst in de dataset die we in 'De variabele.' hebben aangemaakt. De uitkomsten worden per wedstrijd uitgeprint, na elk seizoen is er een streep gezet om het leesbaar te houden, na elke seizoen zijn de punten die de teams gehaald hebben te zien. Aan het einde van de simulatie, nadat alle seizoenen 'gespeelt' zijn, zijn de totaal scores te zien. Dat zijn de scores per team van elk seizoen bij elkaar opgeteld.

```python
for i in range(1000):
    score_ajax = 0
    score_feyenoord = 0
    score_psv = 0
    score_fcUtrecht = 0
    score_willemII = 0
    count = 0
    print('----------')
    for j in range(len(alle_wedstrijden)):
        wedstrijd = alle_wedstrijden[j]
        for i in range(len(alle_wedstrijden[0])):
            count += 1
            getal = random.randint(1, 100)
            #print(count)
            #print(getal)
            thuis = alle_wedstrijden[j][i]
            win_thuis = getal <= thuis[0]
            gelijkspel = thuis[0] < getal <= (thuis[0] + thuis[1])
            verlies_thuis = getal > (thuis[0] + thuis[1])

            if win_thuis and count < 5:
                score_ajax += 3
                totaal_score_ajax += 3
                print('Ajax wint')
            if win_thuis and 4 < count < 9:
                score_ajax == score_ajax
                totaal_score_ajax == totaal_score_ajax
                print('Ajax verliest')
            if gelijkspel and count < 9:
                score_ajax += 1
                totaal_score_ajax += 1
                print('gelijkspel')
            if verlies_thuis and 4 < count < 9:
                score_ajax += 3
                totaal_score_ajax += 3
                print('Ajax wint')
            if verlies_thuis and count < 5:
                score_ajax == score_ajax
                totaal_score_ajax == totaal_score_ajax
                print('Ajax verliest')

            if win_thuis and 8 < count < 13:
                score_feyenoord += 3
                totaal_score_feyenoord += 3
                print('Feyenoord wint')
            if win_thuis and 12 < count < 17:
                score_feyenoord == score_feyenoord
                totaal_score_feyenoord == totaal_score_feyenoord
                print('Feyenoord verliest')
            if gelijkspel and 8 < count < 17:
                score_feyenoord += 1
                totaal_score_feyenoord += 1
                print('gelijkspel')
            if verlies_thuis and 12 < count < 17:
                score_feyenoord += 3
                totaal_score_feyenoord += 3
                print('Feyenoord wint')
            if verlies_thuis and 8 < count < 13:
                score_feyenoord == score_feyenoord
                totaal_score_feyenoord == totaal_score_feyenoord
                print('Feyenoord verliest')  

            if win_thuis and 16 < count < 21:
                score_psv += 3
                totaal_score_psv += 3
                print('Psv wint')
            if win_thuis and 20 < count < 25:
                score_psv == score_psv
                totaal_score_psv == totaal_score_psv
                print('Psv verliest')
            if gelijkspel and 16 < count < 25:
                score_psv += 1
                totaal_score_psv += 1
                print('gelijkspel')
            if verlies_thuis and 20 < count < 25:
                score_psv += 3
                totaal_score_psv += 3
                print('Psv wint')
            if verlies_thuis and 16 < count < 21:
                score_psv == score_psv
                totaal_score_psv == totaal_score_psv
                print('Psv verliest')

            if win_thuis and 24 < count < 29:
                score_fcUtrecht += 3
                totaal_score_fcUtrecht += 3
                print('Fc Utrecht wint')
            if win_thuis and 28 < count < 33:
                score_fcUtrecht == score_fcUtrecht
                totaal_score_fcUtrecht == totaal_score_fcUtrecht
                print('Fc Utrecht verliest')
            if gelijkspel and 24 < count < 33:
                score_fcUtrecht += 1
                totaal_score_fcUtrecht += 1
                print('gelijkspel')
            if verlies_thuis and 28 < count < 33:
                score_fcUtrecht += 3
                totaal_score_fcUtrecht += 3
                print('Fc Utrecht wint')
            if verlies_thuis and 24 < count < 29:
                score_fcUtrecht == score_fcUtrecht
                totaal_score_fcUtrecht == totaal_score_fcUtrecht
                print('Fc Utrecht verliest')

            if win_thuis and 32 < count < 37:
                score_willemII += 3
                totaal_score_willemII += 3
                print('Willem II wint')
            if win_thuis and 36 < count < 41:
                score_willemII == score_willemII
                totaal_score_willemII == totaal_score_willemII
                print('Willem II verliest')
            if gelijkspel and 32 < count < 41:
                score_willemII += 1
                totaal_score_willemII += 1
                print('gelijkspel')
            if verlies_thuis and 36 < count < 41:
                score_willemII += 3
                totaal_score_willemII += 3
                print('Willem II wint')
            if verlies_thuis and 32 < count < 37:
                score_willemII == score_willemII
                totaal_score_willemII == totaal_score_willemII
                print('Willem II verliest')

    print('Seizoen resultaat: Ajax {}, Feyenoord {}, Psv {}, Fc Utrecht {}, Willem II {}'.format(score_ajax, score_feyenoord, score_psv, score_fcUtrecht, score_willemII))
    
    Scores = sorted({"Ajax": score_ajax, "Feyenoord": score_feyenoord, "PSV": score_psv, "FC Utrecht": score_fcUtrecht, "Willem II": score_willemII}.items(), key=lambda kv: (kv[1], kv[0]), reverse=True)
    ranking = []
    for item in Scores:
        ranking.append(item[0])
    df = df.append({'Eerste': ranking[0], 'Tweede': ranking[1],
                    'Derde': ranking[2], 'Vierde': ranking[3],
                    'Vijfde': ranking[4], }, ignore_index=True)
print('Scores van alle seizoenen bij elkaar: Ajax {}, Feyenoord {}, Psv {}, Fc Utrecht {}, Willem II {}'.format(totaal_score_ajax, totaal_score_feyenoord, totaal_score_psv, totaal_score_fcUtrecht, totaal_score_willemII))
```
